##Modelagem Inicial

Primeiramente, se faz necessário definirmos inicialmente todos os estados possíveis, pois eles serão fundamentais para a definição da matriz de transição na qual iremos trabalhar.

Como dito no enunciado, as casas são enumeradas de $0$ até $39$, nos fornecendo uma modelagem de $40$ casos possíveis. Porém, esse raciocínio se demonstra simplista, pois é incapaz de abordar a possibilidade do jogador ser preso e, de acordo com as regras, poder ficar no máximo 3 rodadas e ser solto. Portanto, para realizar tal abordagem, se faz necessário a definição de dois casos artificiais para cada estado. Por exemplo, para uma casa dentro do conjunto $X$, teremos os casos artificiais $X'$ e $X''$, totalizando dessa maneira cerca de $120$ estados. No entanto, ainda falta levar em consideração o máximo de 3 rodadas que o jogador pode ficar na prisão, totalizando mais 3 estados possíveis. 

Em outras palavras, teremos as seguintes definições com relacão aos conjuntos descritos acima:

$$X = [0, 1, 2, ..., 39]$$
$$X' = [40, 41, 42, ..., 79]$$
$$X'' = [80, 81, 82, ..., 119]$$
$$Pris\tilde{a}o = [120, 121, 122]$$

Logo, nossa matriz de transição terá tamanho final de $123$ $ X $ $123$, se assemelhando ao formato abaixo:

$$\begin{equation*}
P = 
\begin{bmatrix}
p_{0} & p_{0} & \cdots & p_{39} & p_{40} & \cdots & p_{79} & x_{80} & \cdots & p_{119} & p_{120} & p_{121} & p_{122}\\
\vdots  & \vdots  & \ddots & \vdots & \vdots & \ddots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots& \vdots\\
p_{0} & p_{0} & \cdots & p_{39} & p_{40} & \cdots & p_{79} & p_{80} & \cdots & p_{119} & p_{120} & p_{121} &p_{122}\\
\vdots  & \vdots  & \ddots & \vdots & \vdots & \ddots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots & \vdots\\
p_{0} & p_{0} & \cdots & p_{39} & p_{40} & \cdots & p_{79} & p_{80} & \cdots & p_{119} & p_{120} & p_{121} & p_{122}
\end{bmatrix}
\end{equation*}$$

Onde cada linha representa, respectivamente, as probabilidades de transição a partir dos estados $p_{0}$ até $p_{39}$, de $p_{40}$ até $p_{79}$, de $p_{80}$ até $p_{119}$ e, por fim, as casas $p_{120}$, $p_{121}$, $p_{122}$.

Como demonstrado nos slides disponibilizados para a elaboração desse projeto, temos que, por definição, uma Cadeia de Markov pode ser representada da seguinte forma:

$$P(X_{t+1} = x_{t+1} | X_{0} = x_{0}, X_{1} = x_{1}, ..., X_{t} = x_{t}) = P(X_{t+1} = x_{t+1} | X_{t} = x_{t})$$

Enquanto isso, na modelagem do Banco Imobiliário, temos o desenho de uma Cadeia de Markov, já que a próxima posição do jogador no tabuleiro é definida por todos os resultados anteriores dos dados, e consequentemente, a posição seguinte do jogador depende de sua posição atual.




Com a matriz de transição estando pré-definida, temos que o estado inicial dela pode ser escrito como: $p_{0} = [1, 0, 0, ..., 0]$, já que, seguindo as regras do jogo, o jogador sempre começará da primeira casa.

Continuamente, agora temos que descobrir como será o resto da matriz e como ela deverá ser preenchida. Começando pela elaboração de conceitos básicos, porém cruciais, de nossa modelagem, temos o conceito de ir para a prisão. Para tal, temos que o jogador, para ir para a prisão, é necessário que as duas faces voltadas para cima dos dois dados sejam iguais no i-ésimo lançamento em uma sequência de até 3 jogadas. Em outras palavras, caso o jogador, no seu primeiro lançamento, tire dados iguais, ele irá para o estado $X'$ tendo o direito de realizar um segundo lançamento. Novamente, caso o lançamento resulte em dois dados com valores repetidos, ele se deslocará para o estado $X''$, conquistando mais uma vez a oportunidade de realizar outro lançamento. Mais uma vez, caso volte a tirar valores repetidos, ele enfim estará na prisão.

Com isso em mente, seja $A$ o evento de "As duas faces voltadas para cima dos dois dados são iguais no lançamento" e $B$ como "ir preso" (seguindo o padrão descrito no enunciado), temos que:

$$P(A) = (\dfrac{6}{36}) = (\dfrac{1}{6})$$

Dessa forma, podemos concluir que: $P(X') = P(X'') = P(A)$

Portanto, como o jogador só pode ir preso após 3 lançamentos, $B$ será igual a:

$$P(B) = P(A) \cdot P(A) \cdot P(A)$$
$$P(B) = (\dfrac{1}{6}) \cdot (\dfrac{1}{6}) \cdot (\dfrac{1}{6}) = (\dfrac{1}{216})$$

Dando continuidade, a probabilidade de sair da prisão pode ser entendida como:

$$p_{120} = P^c(A)$$

$$p_{121} = P^c(A)$$

Isto é:

$$p_{120} = 1 - \dfrac{1}{6}$$

$$p_{121} = 1 - \dfrac{1}{6}$$

Vale ressaltar que quando o jogador está na casa $120$ ou $121$, ele pode apenas se mover para a casa seguinte, ou ir direto para uma casa dentro do conjunto $X$, ou seja, sair da prisão. Além disso, segundo as regras descritas, o jogador que está na casa $122$, de maneira independente de qualquer evento, vai para uma casa dentro do conjunto $X$, visto que já se passaram as 3 rodadas. Logo, 
$$p_{122} = 1$$

Agora, temos que os casos que ainda faltam nessa modelagem tratam das possibilidades do jogador não ir para a prisão. Dessa maneira, podemos definir como $C_{i}$ o evento "jogador anda $i$ casas" a partir da casa em que o mesmo se encontra no início da rodada. Como estamos falando do lançamento de dois dados, temos que i está compreendido no seguinte intervalo:

$$2 ≤ i ≤ 12$$

Dessa forma, basta calcularmos as probabilidades individuais de cada $C_{i}$ com base nos conjuntos possíveis que podem ser observadas a cada lançamento, que podem ser escritos como $Conj_{i}$. Eles, por sua vez, serão descritos logo abaixo. Assim:

$$Conj_{2} = [(1 ,1)]$$
$$Conj_{3} = [(1 ,2), (2, 1)]$$
$$Conj_{4} = [(1 ,3), (3, 1), (2, 2)]$$
$$Conj_{5} = [(1, 4), (4, 1), (2 ,3), (3, 2)]$$
$$Conj_{6} = [(1, 5), (5, 1), (2 ,4), (4, 2), (3, 3)]$$
$$Conj_{7} = [(2, 5), (5, 2), (3 ,4), (4, 3),  (1, 6), (6, 1)]$$
$$Conj_{8} = [(2 ,6), (6, 2), (3, 5), (5, 3), (4, 4)]$$
$$Conj_{9} = [(3 ,6), (6, 3), (4, 5), (5, 4)]$$
$$Conj_{10} = [(4 ,6), (6, 4), (5, 5)]$$
$$Conj_{11} = [(5 ,6), (6, 5)]$$
$$Conj_{12} = [(6 ,6)]$$


Agora que temos os conjuntos de possibilidades mapeadas, basta definirmos $C_{i}$ levando em consideração que o tamanho de nosso espaço amostral é definido como $36$ (número total dos conjuntos mapeados, ou seja, ($6 ⋅ 6$). Logo,

$$C_{2} = \dfrac{1}{36}, C_{3} = \dfrac{2}{36}, C_{4} = \dfrac{3}{36}$$
$$C_{5} = \dfrac{4}{36}, C_{6} = \dfrac{5}{36}, C_{7} = \dfrac{6}{36}$$
$$C_{8} = \dfrac{5}{36}, C_{9} = \dfrac{4}{36}, C_{10} = \dfrac{3}{36}$$
$$C_{11} = \dfrac{2}{36}, C_{12} = \dfrac{1}{36}$$

Como podemos perceber, temos as seguintes equivalências:

$$C_{2} = C_{12}, C_{3} = C_{11}, C_{4} = C_{10},$$
$$C_{5} = C_{9}, C_{6} = C_{8}$$


Porém, em nossa modelagem é possível observar uma incongruência, pois temos que ainda faltam nessa modelagem abordar as possibilidades do jogador não ir para a prisão. Utilizando os conceitos aprendidos no decorrer da disciplina "Números Inteiros e Criptografia RSA", podemos manusear esses casos de modo bastante particular. Portanto, se $i$ é ímpar, não existe possibilidade do jogador ir para a prisão, já que é impossível matemáticamente, pois tal resultado necessariamente implica que as duas faces são diferentes, já que a soma de dois números inteiros resulta obrigatoriamente em um número par. Assim, teremos a seguinte equivalência, onde $Di$ significa "a probabilidade da soma das faces dos dados resultar no valor i", que fora achada anteriormente:

<center>$P(C_{i}) = P(D_{i})$, onde $i$ é ímpar</center>

Por consequência, quando $i$ é par, existe a possibilidade do jogador ir para a prisão. Para tal, devemos considerar a probabilidade encontrada anteriormente para os números pares, ou seja, $C_{2}, C_{4}, C_{6}, C_{8}, C_{10}, C_{12}$ e recalcularmos aplicando as chances de que esses lançamentos possam acabar levando a prisão.

Para isso, basta aplicarmos a seguinte fórmula:

<center>$P(C_{i}) = D_{i} ⋅ \dfrac{possibilidadesSemFacesIguais}{qtdTotal}$, onde $i$ é par</center>

Em outras palavras, $possibilidadesSemFacesIguais$ se refere as chances que temos de sair o número $i$ sem ser através de duas faces iguais e $qtdTotal$ diz respeito a quantidade total de possibilidades que a soma $i$ pode existir. Assim sendo, teremos novas definições e, abaixo, demonstrarei os cálculos que levarão ao novos resultados. Dessa forma:

$$C_{2} = C_{12} = \dfrac{1}{36} ⋅ 0 = 0$$

$$C_{4} = C_{10} = \dfrac{3}{36} ⋅ \dfrac{2}{3} = \dfrac{1}{18}$$

$$C_{6} = C_{8} = \dfrac{5}{36} ⋅ \dfrac{4}{5} = \dfrac{1}{9}$$

Para validar nossa modelagem, basta somarmamos todas as possibilidades possíveis definidas para $C_{i}$ com a chance que o jogador tem de mudar de estado quando tira dados com faces iguais, que é de $\dfrac{6}{36} = \dfrac{1}{6}$. Assim, basta olharmos a conta abaixo:

$$∑C_{i} = \dfrac{5}{6}$$

$$Total = \dfrac{5}{6} + \dfrac{1}{6} = \dfrac{6}{6} = 1$$

Finalizando o nosso processo de modelar o jogo Banco Imobiliário, podemos dizer que as demais probabilidades que não foram definidas no decorrer dessa explicação serão iguais a $0$. Portanto, a matriz de transição pode ser representada como:

```python
import numpy as np
import pandas as pd

def monta_colunas(tamanho):
  colunas = []
  for i in range(0, tamanho):
    colunas.append("p" + str(i))
  return colunas

def monta_linhas(tamanho):
  linhas = []
  for i in range(0, tamanho):
    linhas.append("p" + str(i))
  return linhas

def exibe_matriz(matriz_transicao, tamanho, max_size = 150):
  df = pd.DataFrame(matriz_transicao, columns = monta_colunas(tamanho), index = monta_linhas(tamanho))
  pd.set_option('display.max_columns', max_size)
  pd.set_option('display.max_rows', max_size)
  display(df)

num_casas = 40
num_categoria_estados = 3
num_estados_prisao = 3
num_estados = num_casas * num_categoria_estados + num_estados_prisao
soma_total_dado = 12

p = np.zeros((num_estados, num_estados))

for i in range(num_estados):
  linha = np.zeros(num_estados)
  linha[(i + 2) % num_estados] = 0
  linha[(i + 3) % num_estados] = 2/36
  linha[(i + 4) % num_estados] = 1/18
  linha[(i + 5) % num_estados] = 1/9
  linha[(i + 6) % num_estados] = 1/8
  linha[(i + 7) % num_estados] = 1/6
  linha[(i + 8) % num_estados] = 1/8
  linha[(i + 9) % num_estados] = 1/9
  linha[(i + 10) % num_estados] = 1/18
  linha[(i + 11) % num_estados] = 2/36
  linha[(i + 12) % num_estados] = 0
  if(i < 40):
    linha[120] = 1/6
  elif(i >= 40 and i < 80):
    linha[121] = 1/6
  elif(i >= 80 and i < 123):
    linha[122] = 1/6
  p[i] = linha

exibe_matriz(p, num_estados)
```

###Abaixo, se encontra a resposta para o item D)

O comportamento descrito anteriormente no item C pode ser explicado através da Lei dos Grandes Números (LGN). Através dela, temos que, quanto mais tentativas são realizadas, mais a probabilidade da média aritmética dos resultados observados se aproximará da probabilidade real. Escrevendo de maneira mais formal, seja $X$ a nossa variável aleatória e $n$ a quantidade de casos, podemos chegar a seguinte definição para o seu respectivo valor médio:

<center>$\overline{X_{n}} = E(X)$ quando $n → ∞$</center>

Onde, vale lembrar $E(x)$ é o nosso valor esperado, ou seja, a esperança da variável aleatória $X$.

Complementando, a LGN está intimimente ligada à Distribuição de Bernoulli, pois a ocorrência de um determinado evento onde seus possíveis resultados podem ser resumidos a "sucesso" ou "fracasso", convergirá rumo a sua probabilidade teórica. Na modelagem desenvolvida para esse projeto, podemos definir o ato do "jogador não ir para a cadeia" como sucesso e a o ato do "jogador ser preso" como fracasso. Por exemplo, pegando um caso mais trivial, vamos supor que $M$ seja o resultado do lançamento de uma moeda, podemos representar a LGN através da seguinte fórmula:

<center>$\overline{M_{n}} = \dfrac{m_{1} + m_{2} + m_{3} + ... + m_{n}}{n}= E(M)$ quando $n → ∞$</center>

Logo, a medida que mais testes são realizados, $M$ vai ficando cada vez mais próximo de $\dfrac{1}{2}$.

###Abaixo, se encontra a resposta para o item E)
Ao analisarmos a função massa de probabilidade, temos que a casa em que o jogador passa mais tempo é a prisão, ou seja, a casa $20$. Isso pode ser explicado devido as regras que compôem o jogo. Em absolutamente $todas$ as rodadas é possível enxergar a possibilidade do jogador parar na casa $20$, isto é, ir para a prisão. Além disso, o jogador pode simplesmente "visitar" a prisão, ou seja, após realizar um lançamento de dados cujo o resultado sejam faces distintas, o resultado pode acabar fazendo com que o jogador tenha que caminhar até a casa $20$. Como forma de embasar o argumento aqui utilizado, abaixo é possível visualizar o código de um programa que simula milhares de interações do jogo Banco Imobiliário. Em outras palavras, de modo similar a análise do poema Eugene Onegin, as interações do jogo são armazenadas e, com isso, analiso a probabilidade de cada casa, exibindo-as em tela.

```python
from random import randint
import numpy as np

def calcula_probabilidade(lista_casas, interacoes):
  for i in range (len(lista_casas)):
    if(i % 3 == 0):
      print("\n")
    lista_casas[i] = (lista_casas[i] / interacoes) * 100
    print("Casa: ", i, "\tProbabilidade: {:.2f}".format(lista_casas[i]), "\t", end="")
  maior_probabilidade = max(lista_casas, key=float)
  casa_mais_provavel = np.where(lista_casas == maior_probabilidade)
  print("\n\nCom isso, temos que a casa com a maior probabilidade é a casa:\t", int(casa_mais_provavel[0]))

def lancamento():
  dado_um = randint(1, 6)
  dado_dois = randint(1, 6)
  return dado_um, dado_dois

def monopoly_probabilidades (interacoes):
  cadeia = False
  rodadas_na_cadeia = 0
  posicao_dado = 0
  lista_casas = np.zeros (40)
  for rodada in range(1, interacoes + 1):
    soma_das_faces = 0
    if(cadeia):
      rodadas_na_cadeia = rodadas_na_cadeia + 1
      if(rodadas_na_cadeia == 3):
        cadeia = False
      else:
        dado_um, dado_dois = lancamento()
        if(dado_um != dado_dois):
          cadeia = False
    else:
      dado_um, dado_dois = lancamento()
      if(dado_um == dado_dois):
        dado_um, dado_dois = lancamento()
        if(dado_um == dado_dois):
          dado_um, dado_dois = lancamento()
          if(dado_um == dado_dois):
            posicao_dado = 20
            cadeia = True
      if(not cadeia):
        soma_das_faces = dado_um + dado_dois
    posicao_dado += soma_das_faces
    posicao_dado = posicao_dado % 40
    lista_casas[posicao_dado] += 1
  calcula_probabilidade(lista_casas, interacoes)

monopoly_probabilidades(10000)
```

Assim sendo, através do programa acima podemos ver o pleno compartamento da função massa de probabilidade e como ela é distribuída casa-a-casa, ficando provado que a casa onde o jogador acaba ficando mais tempo é a casa $20$.
